```python
# Cell [7] - Simple version without function
import pandas as pd

# Read the CSV file
df = pd.read_csv("fii_dii_data_sep23.csv")

# Convert date column to datetime
df['date'] = pd.to_datetime(df['date'])

# Sort by date and institution
df = df.sort_values(['date', 'institution', 'trade_type'])

print("Raw data shape:", df.shape)
print("Date range:", df['date'].min(), "to", df['date'].max())
print("Trade types:", df['trade_type'].unique())
print("Institutions:", df['institution'].unique())
print("\nSample data:")
print(df.head(10))
```

```python
# Cell [8] - Explore the data structure
print("=== DATA EXPLORATION ===")
print(f"Data shape: {df.shape}")
print(f"Date range: {df['date'].min()} to {df['date'].max()}")
print(f"Number of days: {df['date'].nunique()}")

print("\n=== TRADE TYPES ===")
print("Available trade types:")
for trade_type in df['trade_type'].unique():
    count = df[df['trade_type'] == trade_type].shape[0]
    print(f"  {trade_type}: {count} records")

print("\n=== INSTITUTIONS ===")
print("Available institutions:")
for inst in df['institution'].unique():
    count = df[df['institution'] == inst].shape[0]
    print(f"  {inst}: {count} records")

print("\n=== SAMPLE DATA BY TRADE TYPE ===")
for trade_type in df['trade_type'].unique():
    print(f"\n--- {trade_type} Sample ---")
    sample = df[df['trade_type'] == trade_type].head(3)
    print(sample[['date', 'institution', 'net_oi']])
```

```python
# Cell [9] - Calculate Options Net OI (CALL - PUT)
print("=== PROCESSING OPTIONS DATA ===")

# Check if we have CALL and PUT data
has_call = 'CALL' in df['trade_type'].unique()
has_put = 'PUT' in df['trade_type'].unique()

print(f"Has CALL data: {has_call}")
print(f"Has PUT data: {has_put}")

if has_call and has_put:
    # Filter options data
    options_df = df[df['trade_type'].isin(['CALL', 'PUT'])].copy()
    print(f"\nOptions data shape: {options_df.shape}")
    
    # Pivot to get CALL and PUT in separate columns
    options_pivot = options_df.pivot_table(
        index=['date', 'institution'], 
        columns='trade_type', 
        values='net_oi',  # Fixed: NET_OI -> net_oi
        fill_value=0
    ).reset_index()
    
    # Calculate Options Net OI = CALL - PUT
    options_pivot['OPTIONS_NET_OI'] = options_pivot['CALL'] - options_pivot['PUT']
    
    # Rename for clarity
    options_pivot = options_pivot.rename(columns={
        'CALL': 'CALL_OI',
        'PUT': 'PUT_OI'
    })
    
    print("\n=== OPTIONS NET OI CALCULATED ===")
    print("Sample results:")
    print(options_pivot.head())
    
else:
    print("Missing CALL or PUT data - creating empty options dataframe")
    options_pivot = pd.DataFrame(columns=['date', 'institution', 'CALL_OI', 'PUT_OI', 'OPTIONS_NET_OI'])
```

```python
# Cell [10] - Process other trade types (CASH, FUTURE-INDEX, etc.)
print("=== PROCESSING OTHER TRADE TYPES ===")

# Get non-options trade types
other_types = [t for t in df['trade_type'].unique() if t not in ['CALL', 'PUT']]
print(f"Other trade types: {other_types}")

if other_types:
    # Filter non-options data
    other_df = df[df['trade_type'].isin(other_types)].copy()
    
    # Pivot to get each trade type as a separate column
    other_pivot = other_df.pivot_table(
        index=['date', 'institution'],
        columns='trade_type', 
        values='net_oi',  # Fixed: NET_OI -> net_oi
        fill_value=0
    ).reset_index()
    
    # Clean column names (replace special characters)
    other_pivot.columns = [col.replace('-', '_') if col not in ['date', 'institution'] else col 
                          for col in other_pivot.columns]
    
    print("\n=== OTHER TRADE TYPES PROCESSED ===")
    print("Sample results:")
    print(other_pivot.head())
    
else:
    print("No other trade types found")
    other_pivot = pd.DataFrame(columns=['date', 'institution'])
```

```python
# Cell [11] - Combine options and other trade types
print("=== COMBINING ALL DATA ===")

# Merge options data with other trade types
if not options_pivot.empty and not other_pivot.empty:
    # Merge both datasets
    final_df = pd.merge(
        other_pivot, 
        options_pivot, 
        on=['date', 'institution'], 
        how='outer'
    ).fillna(0)
    
elif not options_pivot.empty:
    # Only options data available
    final_df = options_pivot.copy()
    
elif not other_pivot.empty:
    # Only other trade types available
    final_df = other_pivot.copy()
    # Add empty options columns
    final_df['CALL_OI'] = 0
    final_df['PUT_OI'] = 0
    final_df['OPTIONS_NET_OI'] = 0
    
else:
    print("No data to combine!")
    final_df = pd.DataFrame()

if not final_df.empty:
    # Sort by date and institution
    final_df = final_df.sort_values(['date', 'institution'])
    
    print(f"Combined data shape: {final_df.shape}")
    print("Columns in final dataset:")
    print(final_df.columns.tolist())
    
    print("\n=== SAMPLE OF COMBINED DATA ===")
    print(final_df.head())
else:
    print("No data available for processing")
```

```python
# Cell [12] - Calculate daily changes for each metric
print("=== CALCULATING DAILY CHANGES ===")

if not final_df.empty:
    # Get all numeric columns (excluding date and institution)
    numeric_columns = final_df.select_dtypes(include=[np.number]).columns.tolist()
    print(f"Numeric columns for change calculation: {numeric_columns}")
    
    # Calculate daily changes for each institution
    for col in numeric_columns:
        # Calculate change = today's value - yesterday's value
        final_df[f'{col}_CHANGE'] = final_df.groupby('institution')[col].diff()
    
    # Show columns after adding changes
    print(f"\nColumns after adding changes: {final_df.columns.tolist()}")
    
    print("\n=== SAMPLE WITH DAILY CHANGES ===")
    print("First few rows (will have NaN for first day changes):")
    print(final_df.head(8))
    
    print("\n=== RECENT DATA WITH CHANGES ===")
    print("Last few rows:")
    print(final_df.tail(8))
    
else:
    print("No data available for change calculation")
    
```

```python
# Cell [13] - Save to CSV and show summary
print("=== SAVING RESULTS ===")

if not final_df.empty:
    # Save to CSV
    output_file = 'processed_fii_dii_data.csv'
    final_df.to_csv(output_file, index=False)
    print(f"Data saved to: {output_file}")
    
    print("\n=== SUMMARY STATISTICS ===")
    # Get position columns (exclude change columns for summary)
    position_cols = [col for col in final_df.columns 
                    if not col.endswith('_CHANGE') and col not in ['date', 'institution']]
    
    if position_cols:
        print("Summary of positions by institution:")
        summary = final_df.groupby('institution')[position_cols].agg(['mean', 'std']).round(2)
        print(summary)
    
    print("\n=== LATEST POSITIONS ===")
    latest_date = final_df['date'].max()
    latest_data = final_df[final_df['date'] == latest_date]
    print(f"Latest date: {latest_date}")
    print(latest_data[['institution'] + position_cols])
    
    print(f"\n=== SUCCESS! ===")
    print(f"Final dataset shape: {final_df.shape}")
    print(f"Date range: {final_df['date'].min()} to {final_df['date'].max()}")
    
else:
    print("No data to save")
```
